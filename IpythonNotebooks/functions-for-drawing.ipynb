```python
import networkx as nx
%matplotlib inline
import matplotlib.pyplot as plt
```

### defining function

```python
def draw_graph():
    G = nx.Graph()
    G.add_edges_from([(1,2),(2,3),(1,4),(1,3)])
    nx.draw(G,node_size=800,node_color='red',with_labels=True)
```

```python
draw_graph()
```

```python
def draw_barbasi_albert_graph():
    G = nx.barabasi_albert_graph(100,10)
    hst = nx.degree_histogram(G)
    plt.hist(hst,bins=20,color='red')
```

```python
draw_barbasi_albert_graph()
```

### Graph g= {(A,B),(B,C),(C,A)} each tuple is an edge

```python
def simple_graph():
    G = nx.Graph()
    
    G.add_edge('A','B')
    G.add_edge('B','C')
    G.add_edge('C','A')
    
    nx.draw(G,with_labels=True,node_size=800,node_color='b')
```

```python
simple_graph()
```

## deleting nodes using function

```python
def delete_nodes():
    G = nx.Graph()
    G.add_edge('A','B')
    G.add_edge('B','C')
    G.add_edge('C','A')
    
    nx.draw(G,node_size=800,with_labels=True,node_color='y')
    print('removing node B')
    G.remove_node('B')
    
    nx.draw(G,node_size=800,with_labels=True,node_color='r')
```

```python
delete_nodes()
```

## Deleting specific edges from graph

```python
def delete_edges():
    G = nx.Graph()
    G.add_edge('A','B')
    G.add_edge('B','C')
    G.add_edge('C','A')
    
    nx.draw(G,node_size=800,with_labels=True,node_color='y')
    plt.show()
    
    # removing edges
    
    G.remove_edge('A','B')
    G.remove_edge('B','C')
    
    nx.draw(G,node_size=800,with_labels=True,node_color='y')
    plt.show()
```

```python
delete_edges()
```

### Customize node color,edge color and node size of Graph

```python
def node_size_and_color():
    G = nx.Graph()
    
    G.add_edge('A','B')
    G.add_edge('B','C')
    G.add_edge('C','A')
    
    nx.draw(G,with_labels=True)
    plt.show()
    
    node_lst=['A','B','C']
    
    colors=['c','m','y']
    
    nodesize = [900,1000,1200]
    
    edg_clr=['c','r','b']
    
    edg_lst=[('A','B'),('B','C'),('C','A')]
    
    nx.draw(G,with_labels=True,node_size=nodesize,nodelist=node_lst,node_color=colors,edgelist=edg_lst,edge_color=edg_clr)
```

```python
node_size_and_color()
```
