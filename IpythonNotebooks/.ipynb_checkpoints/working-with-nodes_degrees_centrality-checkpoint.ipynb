## nodes, degrees centrality

```python
from networkx import *
import sys
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
n = 10 # nodes
m = 20 # edges

g = nx.gnm_random_graph(n,m)

print('Nodes degress clustering')

for v in nodes(g):
    print('%s %d %f '%(v,degree(g,v),clustering(g,v)))
```

```python
nx.draw(g,with_labels=True)
```
