```python
import networkx as nx
%matplotlib inline
import matplotlib.pyplot as plt
```

## Classical graphs

### barbel , lollipop, Binomial , Chordal cycle , Complete Bipartite graph

```python
b = nx.barbell_graph(10,10)

nx.draw(b,node_size= 800, with_labels=True,node_color='r')
```

```python
print(nx.info(b))
```

```python
l = nx.lollipop_graph(10,20)

nx.draw(l,node_size= 800, with_labels=True,node_color='c')
```

```python
print(nx.info(l))
```

```python
plt.figure(figsize=(10,7))
bi = nx.binomial_graph(100,0.5)

nx.draw(bi,node_size= 650, with_labels=True,node_color='g')
```

```python
plt.figure(figsize=(10,7))
bi = nx.binomial_graph(100,0.15)

nx.draw(bi,node_size= 650, with_labels=True,node_color='g')
```

```python
print(nx.info(bi))
```

```python
c = nx.chordal_cycle_graph(20)

nx.draw(c,node_size= 650, with_labels=True,node_color='tomato')
```

```python
print(nx.info(c))
```

```python
cb = nx.complete_bipartite_graph(3,5)

nx.draw(cb,node_size= 650, with_labels=True,node_color='orange')
```

```python
print(nx.info(cb))
```
