## Small famous graphs

```python
import networkx as nx
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
p = nx.petersen_graph()

nx.draw(p,node_size=800,with_labels=True)
```

```python
t = nx.tutte_graph()

nx.draw(t,node_size=600,with_labels=True,node_color='cyan')
```

```python
s = nx.sedgewick_maze_graph()
nx.draw(s,node_size=800,with_labels=True,node_color='g')
```

```python
tt = nx.tetrahedral_graph()

nx.draw(tt,node_size=800,with_labels=True)
```

## Famous Social network graphs

```python
k = nx.karate_club_graph()

nx.draw(k,node_size=600,with_labels=True,node_color='tomato')
```

```python
print( nx.info(k))
```

```python
d = nx.davis_southern_women_graph()

nx.draw(d,node_size=700,with_labels=True,node_color='blue')
```

```python
plt.figure(figsize=(10,7))
d = nx.davis_southern_women_graph()

nx.draw(d,node_size=1000,with_labels=True,node_color='red')
```

```python
print(nx.info(d))
```

```python
print(d.edges)
```

```python
for i in d.edges(data=True):
    print(i)
```

```python
plt.figure(figsize=(10,7))
f = nx.florentine_families_graph()

nx.draw(f,node_size=2000,with_labels=True,node_color='cyan')
```

```python
print(nx.info(f))
```

```python
for i in f.edges(data=True):
    print(i)
```

```python
for i in f.nodes(data=True):
    print(i)
```
