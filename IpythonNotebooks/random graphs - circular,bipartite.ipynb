```python
from networkx import *
import sys
%matplotlib inline
import matplotlib.pyplot as plt
```

## 2d random graph

```python
G = nx.grid_2d_graph(4,4)
pos= nx.spring_layout(G, iterations=100)

plt.subplot(221)
nx.draw(G,pos,font_size=8)

plt.subplot(222)
nx.draw(G,pos,node_size=0,node_color='k')

plt.subplot(223)
nx.draw(G,pos,node_size=1,node_color='b',with_labels=False)

plt.subplot(224)

# converting undirected to direct

H = G.to_directed()
nx.draw(H,pos,node_size=20,node_color='k')

plt.savefig('img/grids.png')

plt.show()
```

### Circular Graphs

```python
C = nx.star_graph(20)

pos = nx.spring_layout(C,iterations=20)

colrs = range(20)

nx.draw(C,node_color='blue',edge_color=colrs,width=4,edge_cmap=plt.cm.Blues,with_labels=True)
```

## Bipartite graphs 
##### useful for modeling relationship between two entities

```python
from networkx.algorithms import bipartite
```

```python
b = nx.Graph()
```

```python
cities= ['bcm','hyd','nzb','ban','kdp','chn']

people=['ss','Student']

#adding nodes 

b.add_nodes_from(cities,bipartite='cities')
b.add_nodes_from(people,bipartite='people')

# define edges
edges = [('ss','bcm'),('Student','hyd'),('ss','kdp'),('Student','nzb'),('ss','chn'),('Student','ban'),('ss','hyd')]

b.add_edges_from(edges)

nx.draw(b,with_labels=True,node_color='c',node_size=700)
```

## House graphs

```python
h = nx.house_graph()
```

```python
pos = {0: (0,0),1:(1,0),2:(0,1),3:(1,1),4:(0.5,2)}
```

```python
nx.draw_networkx_nodes(h,pos,node_size=2000,nodelist=[4])

nx.draw_networkx_nodes(h,pos,node_size=600,nodelist=[0,1,2,3],node_color='c')

nx.draw_networkx_edges(h,pos,alpha=0.5,width=6)

#plt.axis('off')
```
