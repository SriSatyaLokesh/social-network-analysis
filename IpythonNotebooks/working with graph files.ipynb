## writing and reading graph files using pickle

```python
import networkx as nx
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
g = nx.path_graph(6)
nx.draw(g,with_labels=True,node_size=800,node_color='c')
```

## writig files to disk

```python
nx.write_gpickle(g,'graphs/path_graph.gpickle')
```

```python
import os

os.listdir('graphs/')
```

## Reading graph data

```python
data = nx.read_gpickle('graphs/path_graph.gpickle')
```

```python
nx.draw(data,with_labels=True)
```

```python
print(nx.info(data))
```

## Reading and writing Edgelist to a file

```python
#writing
nx.write_edgelist(nx.path_graph(20),'graphs/path_edge_list.txt')
```

```python
import os

os.listdir('graphs/')
```

```python
#reading

g= nx.read_edgelist('graphs/path_edge_list.txt')
nx.draw(g,with_labels=True)
```

```python
h = nx.read_edgelist('graphs/path_edge_list.txt',create_using=nx.DiGraph(),nodetype=int)

n,e = h.order() , h.size()

avg = float(e)/n

print('Nodes : ',n)
print('Edges : ',e)
print('Avg degree : ',avg)
```

### Reading files using open function
##### open file -> read from it -> close it

```python
f = open('graphs/path_edge_list.txt','rb')

g = nx.read_edgelist(f)

nx.draw(g,with_labels=True,node_size=800,node_color='#A0E352')

f.close()
```

```python
print(nx.info(g))
```
