```python
import networkx as nx
%matplotlib inline
import matplotlib.pyplot as plt
```

## Plotting circular graphs and saving as images

```python
G = nx.cycle_graph(24)
```

```python
pos = nx.spring_layout(G,iterations=200)
```

```python
nx.draw(G,pos,node_color=range(24),node_size=900,cmap=plt.cm.Blues)

plt.savefig('img/cmap_node.png')
```
